# Hybrid Search Techniques

## 1- Dense Sparse Technique
##### Compining power of 2 searching techniques : 
1. Dense Retrieval -> Semantic search
2. Sparse Retrieval -> Exact keyword search 

- Score(desne)  = Cosine_similarity(Input_Query,Vectore_Store)
- Score(sparse) = TF-IDF           (Input_Query,Vectore_Store)
- alpha = weightage of each of them in the below equation

- **Score(hybrid)** = alpha * Score(dense) + (1 - alpha) * Score(sparse)

## 2- Re-Ranking Technique
##### Second-stage filtering process in retirieval systems,especially in RAG pipelines : 

1. Use a fast retriever (like BM25, FAISS , hybrid) to fetch top-k documents quickly
2. Use a more accurate but slower model (like cross-encoder or LLM) to re-score and reorder those documents bu relevance to the query

## 3- MMR [Maximal Marginal Relevant]
##### MMR is a powerful diversity-aware retririeval technique used in ingo retrieval and RAG to balance relevance and novelty when selecting documents.

**MMR Selects documents that are both :**
1. Relevant to the query
2. Diverse from each other (non-redundant)

- lamda = weightage of each of them in the below equation
- MMR(document) = lamda * similarity(document,query) - (1 - lamda) * max (sem scores)(similarity(document,query))

#### When to use MMR ?

1. In RAG : To avoid feeging the llm redundunt documents
2. Chatbot : FAQ , Search app , Doc browser
3. Retriever returns many results
4. MMR + Hybrid retrieval 

#### When **NOT** to use MMR ?

1. Extremely Short Context Window
2. You need precisiom only
3. Documents already diverse
4. You're already reranking with LLM
